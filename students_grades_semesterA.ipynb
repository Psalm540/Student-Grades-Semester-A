```python
import sqlite3
import random
from faker import Faker

# Connect to database
import sqlite3
conn = sqlite3.connect(r"C:\Users\Owner\Desktop\student_grades_semesterA.db")
cursor = conn.cursor()

fake = Faker()

# -------------------------
# Insert Students (500)
# -------------------------
for _ in range(500):
    cursor.execute("""
        INSERT INTO Students (full_name, gender, date_of_birth, enrollment_year)
        VALUES (?, ?, ?, ?)
    """, (
        fake.name(),
        random.choice(["Male", "Female"]),
        fake.date_of_birth(minimum_age=18, maximum_age=30),
        random.randint(2021, 2025)
    ))

# -------------------------
# Insert Modules (5)
# -------------------------
modules = [
    ("Machine Learning", 20, "Dr. Smith"),
    ("Data Mining", 15, "Dr. Johnson"),
    ("Statistics for DS", 20, "Dr. Brown"),
    ("Python Programming", 15, "Dr. White"),
    ("Big Data Analytics", 20, "Dr. Green")
]

cursor.executemany("""
    INSERT INTO Modules (module_name, credit_units, lecturer_name)
    VALUES (?, ?, ?)
""", modules)

# -------------------------
# Insert Assessments (3 per module)
# -------------------------
cursor.execute("SELECT module_id FROM Modules")
module_ids = cursor.fetchall()

assessment_types = ["Exam", "Coursework", "Project"]

for module in module_ids:
    for assess in assessment_types:
        cursor.execute("""
            INSERT INTO Assessments (module_id, assessment_type, max_score)
            VALUES (?, ?, ?)
        """, (module[0], assess, 100))

# -------------------------
# Insert Grades (1000+ rows)
# -------------------------
cursor.execute("SELECT student_id FROM Students")
students = cursor.fetchall()

cursor.execute("SELECT assessment_id FROM Assessments")
assessments = cursor.fetchall()

for student in students:
    for assessment in assessments:
        cursor.execute("""
            INSERT INTO Grades (student_id, assessment_id, score)
            VALUES (?, ?, ?)
        """, (
            student[0],
            assessment[0],
            random.randint(30, 100)
        ))

# Commit and close
conn.commit()
conn.close()

print("Database successfully populated with 1000+ records.")
```
